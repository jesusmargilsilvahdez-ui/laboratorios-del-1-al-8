**Contar del 1 al numero interesado**

```python
n = int(input("Ingresa un número entero positivo: "))
contador = 1

while contador <= n:
    print(contador)
    contador += 1
```

**Mostrar solo números pares o impares usando saltos controlados (sin if):**

```python
n = int(input("Hasta qué número par quieres contar? "))
contador = 2
while contador <= n:
    print(contador)
    contador += 2
```

Con for

El bucle for se usa cuando ya sabes cuántas veces quieres repetir algo.
Python recorre una secuencia (como range(), una lista o una cadena).

```python
for i in range(1, 6):
    print(i)
```

range(1, 6) genera los números 1, 2, 3, 4, 5.

En cada vuelta, i toma uno de esos valores.

El for controla automáticamente el contador y la condición.

No necesitas inicializar ni actualizar manualmente la variable.

Con while

El bucle while se usa cuando no sabes exactamente cuántas veces se repetirá, sino que depende de una condición que puede cambiar durante la ejecución.

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

Qué ocurre aquí:

Tú defines la condición de parada (i <= 5).

Tú controlas el incremento o actualización (i += 1).

Si olvidas actualizar i, puedes crear un bucle infinito.

Usa for → cuando repites algo un número conocido de veces.

Usa while → cuando repites algo mientras se cumpla una condición, y no sabes cuántas veces ocurrirá.

```python
palabra = "sol"
for letra in palabra:
    print(letra)
```

En Python, las variables usadas en un for se crean en el momento del bucle, sin necesidad de declaración previa.

 Esto es válido para cualquier variable en Python (no se declaran, solo se asignan).

 Y Python sobreescribe la variable en cada iteración.

```python
palabra = input("Ingresa una palabra: ")
vocales = "aeiou"
contador = 0

for letra in palabra:
    if letra in vocales:
        break
    contador += 1

print("Letras antes de la primera vocal:", contador)
```

```python
palabra = input("Ingresa una palabra: ")
vocales = "aeiou"
contador = 0
i = 0

while i < len(palabra) and palabra[i] not in vocales:
    contador += 1
    i += 1

print("Letras antes de la primera vocal:", contador)
```
